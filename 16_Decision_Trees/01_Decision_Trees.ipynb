###### Feature Engineering

```python
X = pd.get_dummies(df.drop('species',axis=1), drop_first=True)
y = df['species']
```

###### Train| Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=101)
```

###### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(X_train,y_train)
```

```python
y_pred = dt_model.predict(X_test)
```

```python
y_pred
```

###### Evaluation Report

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
dt_model.feature_importances_
```

```python
pd.DataFrame(index=X.columns, data= dt_model.feature_importances_, columns=['Feature Importance'])
```

###### Visualize the Tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8))

plot_tree(dt_model, feature_names=X.columns)

plt.show()
```

###### Exploring Hyperparameters

```python
def report_model(mod):
    mod_pred = mod.predict(X_test)
    print(classification_report(y_test,mod_pred))
    print('\n')
    plt.figure(figsize=(12,8))
    plot_tree(mod, filled=True,feature_names=X.columns)
```

```python
#Max Depth

max_d_model = DecisionTreeClassifier(max_depth=2)
max_d_model.fit(X_train,y_train)
```

```python
report_model(max_d_model)
```

```python
#Max Leaf Nodes

ml_model = DecisionTreeClassifier(max_leaf_nodes=3)
ml_model.fit(X_train,y_train)
```

```python
report_model(ml_model)
```
