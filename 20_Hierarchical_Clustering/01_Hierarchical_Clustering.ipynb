```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
df = pd.read_csv('cluster_mpg.csv')
df = df.dropna()
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies = pd.get_dummies(df.drop('name',axis=1))
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
scaled_data = sc.fit_transform(df_w_dummies)
```

```python
scaled_df = pd.DataFrame(scaled_data,columns = df_w_dummies.columns)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(scaled_df,cmap='magma')
```

```python
sns.clustermap(scaled_df)
```

```python
df_w_dummies.corr()
```

```python
sns.clustermap(scaled_df,row_cluster=False)
```

```python
sns.clustermap(scaled_df,col_cluster=False)
```

###### Scikit-Learn

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
cluster_labels
```

```python
sns.scatterplot(data=df,x='mpg',y='weight',hue=cluster_labels)
```

###### Exploring Number of Clusters with Dendrograms

```python
model1 = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
model1.fit_predict(scaled_df)
```

###### Linkage Model

```python
linkage_matrix = hierarchy.linkage(model1.children_)
linkage_matrix
```

```python
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(linkage_matrix, truncate_mode='lastp', p =48)
```
