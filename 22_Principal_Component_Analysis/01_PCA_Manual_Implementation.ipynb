```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_X = sc.fit_transform(df)
```

```python
scaled_X
```

```python
# Grab covariance matrix
cov_mat = np.cov(scaled_X, rowvar=False)
```

```python
#Get Eigen vectors and eigen values
eigen_values, eigen_vectors = np.linalg.eig(cov_mat)
```

```python
#eigen_values
```

```python
#eigen_vectors
```

```python
#choose number of components
num_components = 2
```

```python
#Get index sorting key based on Eigen Values
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
#Gen num_components of Eignen Values and Eigen Vectors
eigen_values, eigen_vectors = eigen_values[sorted_key], eigen_vectors[:,sorted_key]
```

```python
#projection step of the original points on to Principal Components
principal_components = np.dot(scaled_X,eigen_vectors)
```

```python
principal_components
```

```python
plt.figure(figsize=(12,9))
plt.scatter(x=principal_components[:,0],y=principal_components[:,1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
