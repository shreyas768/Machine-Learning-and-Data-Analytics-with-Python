```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

###### Creating a Figure

```python
#creates blank canvas
fig = plt.figure(figsize=(4,4))

##Note: All the commands related to a particular figure should go in the same cell

#add set of axes to figure
ax = fig.add_axes([0,0,1,1]) #Left, Bottom, Width, Height

#plot on that set of axes
ax.plot(y,x)

plt.show()
```

```python
type(fig)
```

###### Add another set of axes to the figure

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) #larger figure 
ax2 = fig.add_axes([0.15,0.15,0.5,0.5]) #smaller figure

#add a plot outside of the canvas between x,y

ax3 = fig.add_axes([1.2,1.2,0.4,0.4])

#add figure to ax1
ax1.plot(b,a)
ax1.set_xlabel('This is B')
ax1.set_ylabel('This is A')


#add figure to ax2
ax2.plot(a,b)
ax2.set_xlabel('This is A')
ax2.set_title('Title for smaller plot')
ax2.set_xlim(2,8)
ax2.set_ylim(2000,4000)

#add figure to ax3
ax3.plot(x,y)

#save_fig
fig.savefig('demo_oops.png')
plt.show()
```
