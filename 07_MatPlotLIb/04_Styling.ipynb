```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2*x
```

###### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label = 'x squared')
ax.plot(x, y , label = ' x vs y')

ax.legend(loc = (1.1, 0.5))
```

###### Colors, Linewidth, Linestyles etc.

```python
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, color='#556B2F', marker='*', ls=':',lw=2, ms=12)
ax.plot(x, x**3, color='red', marker='>', linestyle='dashed',linewidth=4, markersize=19, mec = 'blue', mew = 7, mfc = 'red')
```
