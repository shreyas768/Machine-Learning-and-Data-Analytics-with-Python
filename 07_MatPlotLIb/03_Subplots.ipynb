```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots()

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
fig, axes = plt.subplots(ncols=2,nrows=1)

for ax in axes:
    ax.plot(x,y)
    x = x *8
    y = y + 1000
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(x,y)

axes[1,1].plot(a,b)

axes[0,1].plot(a,b)
axes[0,1].set_title('this is 0,1 subplot')

axes[1,0].plot(b,a)
axes[1,0].set_ylabel('a')


#Parameters at figure level
fig.suptitle('Fig level title here', fontsize = 23)


plt.tight_layout()
fig.savefig('subplots_demo.png')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].plot(x,y)

axes[1,1].plot(a,b)

axes[0,1].plot(a,b)
axes[0,1].set_title('this is 0,1 subplot')

axes[1,0].plot(b,a,color='red', marker='>', linestyle='dashed',linewidth=4, markersize=19, mec = 'blue', mew = 7, mfc = 'red')
axes[1,0].set_ylabel('a')


#Parameters at figure level
fig.suptitle('Fig level title here', fontsize = 23)


#Manual Spacing
fig.subplots_adjust(hspace=0.5, wspace=0.5)



fig.savefig('subplots_demo.png')
```
