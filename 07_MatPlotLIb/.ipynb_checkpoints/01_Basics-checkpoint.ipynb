```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

###### Using matplotlib with plt.plot() function calls

```python
plt.plot(x,y)
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
plt.title('This is title')
plt.xlim(2,6)
plt.ylim(2,10)
plt.savefig('demo.png')
plt.show()
```
