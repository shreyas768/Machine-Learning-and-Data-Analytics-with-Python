```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs = pd.read_csv('cluster_blobs.csv')
blobs.head(5)
```

```python
sns.scatterplot(data=blobs, x = 'X1', y = 'X2')
```

```python
moons = pd.read_csv('cluster_moons.csv')
moons.head(5)
```

```python
sns.scatterplot(data=moons, x = 'X1', y = 'X2')
```

```python
circles = pd.read_csv('cluster_circles.csv')
circles.head()
```

```python
sns.scatterplot(data=circles, x = 'X1', y = 'X2')
```

```python
def display_category(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1',y='X2', hue = labels, palette='Set1')
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2)
```

```python
display_category(km,moons)
```

```python
km3 = KMeans(n_clusters=3)
display_category(km3,blobs)
```

```python
display_category(km,circles)
```

###### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.6)

display_category(dbscan,blobs)
```

```python
dbscan = DBSCAN(eps=0.25)

display_category(dbscan,moons)
```
