```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mouse_viral_study.csv')
df.head(10)
```

```python
sns.scatterplot(x='Med_1_mL',y='Med_2_mL',data=df, hue = 'Virus Present')

x = np.linspace(0,10,100)
m = -1
b = 11
y = m*x + b

plt.plot(x,y,'k')
```

###### SVM - Support Vector Machine

```python
y = df['Virus Present']
X = df.drop('Virus Present', axis = 1)
```

```python
from sklearn.svm import SVC
```

```python
#help(SVC)
```

```python
model = SVC(kernel='linear',C=1000)
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```

### Hyperparameters

###### C

```python
model = SVC(kernel='linear',C=.05)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

### Kernel

###### RBF

```python
model = SVC(kernel='rbf',C=1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

###### Degree (Polynomial Kernel Only)

```python
model = SVC(kernel = 'poly', C= 1, degree = 1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel = 'poly', C= 1, degree = 5)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

###### Gamma (rbf, polu, sigmoid)

```python
model = SVC(kernel='rbf',C=1, gamma = .1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

###### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1],
             'kernel':['linear','rbf']}
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_params_
```
