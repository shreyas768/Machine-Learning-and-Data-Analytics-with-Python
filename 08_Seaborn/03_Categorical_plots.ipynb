```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

###### Countplot()

```python
plt.figure(figsize=(10,4))
sns.countplot(x='division', data = df, hue = 'level of education', palette='hot')
```

###### barplot()

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education', y = 'salary', data = df, estimator= np.mean,hue ='division')
plt.legend(loc = (1.02, .1))
```

##### BoxPlot

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
df.info()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data =df, x = 'parental level of education', y = 'math score', hue = 'race/ethnicity')
plt.show()
```

```python
plt.figure(figsize=(10,16))
sns.boxplot(data =df, y = 'parental level of education', x = 'math score', hue = 'race/ethnicity', orient='h')
plt.show()
```

###### Violin Plot

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education', y = 'reading score', data = df, hue = 'gender', split = True, 
              inner = 'quartile')
```

###### Swarmplot

```python
sns.swarmplot(x = 'math score', hue = 'gender', data = df, size = 4, y = 'lunch', dodge = True)
```
