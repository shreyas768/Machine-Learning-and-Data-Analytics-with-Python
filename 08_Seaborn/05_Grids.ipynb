```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df
```

#### catplot()

```python
sns.catplot(x = 'gender', y = 'math score', data = df, kind = 'bar', row = 'lunch', 
            col = 'test preparation course')
```

###### FacetGrid

```python
g = sns.FacetGrid(data = df, col='gender', row = 'lunch')
g.map(plt.scatter, 'math score', 'reading score')
```
