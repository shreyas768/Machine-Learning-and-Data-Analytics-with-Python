```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
```

###### Scatterplot

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x = 'salary', y = 'sales')
plt.show()
```

###### Seaborn Parameters

```python
plt.figure(figsize=(10,4))
sns.scatterplot(data=df, x = 'salary', y = 'sales', hue = 'division')
plt.show()
```

```python
df['work experience'].unique()
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(data=df, x = 'salary', y = 'sales', hue = 'work experience', palette= 'ocean')
plt.show()
```

Palettes : https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
plt.figure(figsize=(10,4))
sns.scatterplot(data=df, x = 'salary', y = 'sales', hue = 'work experience', palette= 'ocean', size= 'training level' )
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x = 'salary', y = 'sales', hue = 'work experience', palette= 'ocean',
                size= 'training level', style = 'level of education' )
plt.savefig('example_scatterplot.png')
plt.show()
```
