```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df
```

###### JointPlot

```python
sns.jointplot(x = 'math score', y = 'writing score', data = df)
```

```python
sns.jointplot(x = 'math score', y = 'writing score', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'math score', y = 'writing score', data = df, kind = 'kde')
```

###### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'gender', palette = 'Set2', corner=True, diag_kind= 'hist')
```
