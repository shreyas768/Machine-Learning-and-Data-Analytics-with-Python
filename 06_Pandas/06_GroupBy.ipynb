```python
df_4 = df[(df['horsepower'] != '?') & (df['cylinders']==4)]
new_hp_4 = str(int(np.floor(df_4['horsepower'].astype(int).mean())))
```

```python
df['horsepower'] = df['horsepower'].replace('?',new_hp_4).astype(int)
```

###### Compute average weight of cars released in year 1974 from the U.S (origin 1) .

```python
df = pd.read_csv('mpg.csv')
df_new = df[[ 'weight', 'model_year','origin']]
df_new
```

```python
df_new.groupby(['origin','model_year']).mean().loc[(1,74)].iloc[0]
```

```python
print(f"The average weight of cars released in year 1974 from the U.S is {np.round(df_new.groupby(['origin','model_year']).mean().loc[(1,74)].iloc[0],2)}")
```
