###### 

* total bill < 10 -->> *
* 10 < total bill 20 -->> **
* 20 and 25 -->> ***
* 25 and 30 -->> ****
* total bill > 30 -->> ****

```python
def star_cat(amount):
    if amount < 10:
        return '*'
    elif 10 <= amount < 20:
        return '**'
    elif 20 <= amount < 25:
        return '***'
    elif 25 <= amount < 30:
        return '****'
    else:
        return '*****'
```

```python
df['customer_rating'] = df['total_bill'].apply(star_cat)
```

```python
df.head()
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
df['tip_quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
```
