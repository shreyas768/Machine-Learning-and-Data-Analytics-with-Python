```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

##### Conditional Filtering

```python
bool_val = df['total_bill'] > 20
```

```python
df[bool_val == False]
```

```python
df[bool_val]
```

```python
df[df['total_bill'] > 20]
```

```python
df[df['smoker'] == 'Yes']
```

###### How many non-smoking groups came for Lunch ?

```python
df1 = df[df['smoker'] == 'No']

l = len(df1[df1['time'] == 'Lunch'])

print(f'{l} non-smoking groups came for Lunch')
```

###### Multiple Conditions

* AND -->> &
* OR -->> |
* NOT -->> ~

```python
len(df[(df['smoker'] == 'No') & (df['time'] == 'Lunch')])
```

###### How many Male customer turned up in the restaurant barring Saturdays?

```python
len(df[(df['sex'] == 'Male') & (df['day'] != 'Sat')])
```

```python
len(df[(df['sex'] == 'Male') & ~(df['day'] == 'Sat')])
```

###### Get all the customers' name who turned up on weekends

```python
df[(df['day'] == 'Sat') | (df['day'] == 'Sun')]['Payer Name']
```

###### .isin()

```python
weekend = ['Sat','Sun']
df[df['day'].isin(weekend)]
```
